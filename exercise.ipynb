```python
import pandas as pd
import random as r
import numpy as n
import matplotlib.pyplot as plt
import collections as c
```

```python
#1 Picking 1000 random  no.s bet 60,70 and saving it in list Age[]
Age = []
for i in range(1000):
    Age.append(r.randint(60, 70))
print(Age)
```

```python
#2 Mean of Age[]
mean=n.mean(Age)
print(mean)
```

```python
#2 Std of Age[]
std=n.std(Age)
print(std)
```

```python
#3 plotting histogram of Age[] showing frequency of each age entity
plt.hist(Age)
plt.xlabel("Age")
plt.ylabel("Freq")
```

```python
#4 randomly chosing 30 elements from Age[] and saving it into Sample[]
sample=[]
for i in range(30):
    sample.append(r.choice(Age))
print(sample)
```

```python
#5 mean of Sample[] and adding it to S_mean[]
meansam=n.mean(sample)
print(meansam)
S_mean=[]
S_mean.append(meansam)
```

```python
#6 repeating 4 & 5 for 500 times 
for i in range(500):
    sample=[]
    for i in range(30):
        sample.append(r.choice(Age))
    S_mean.append(n.mean(sample))
print(S_mean)
```

```python
#7 plotting histogram of S_mean
plt.hist(S_mean)
plt.xlabel("Mean_Age")
plt.ylabel("Freq")
```

```python
#8 comparing mean and std of S_mean[] using central limit theorem
S_mean_m=n.mean(S_mean)
S_mean_std=n.std(S_mean)
print(S_mean_m)                # here mean of S.mean[] is aprox. equal to the mean of Age[] database
print(S_mean_std)              # also std of S.mean[] is aprox. equal to (std of Age[])/root(30)
```

```python
#9 Increasing the sample size by 50
S1_mean=[]
for i in range(500):
    sample=[]
    for i in range(50):
        sample.append(r.choice(Age))
    S1_mean.append(n.mean(sample))
print(S1_mean)
```

```python
#9 Histogram plot for S1_mean[]
plt.hist(S1_mean)
```

```python
#10 Repeating 8 to see better results on a larger sample size of 50
S1_mean_m=n.mean(S1_mean)
S1_mean_std=n.std(S1_mean)
print(S1_mean_m)                # here mean of S.mean[] is aprox. equal to the mean of Age[] database
print(S1_mean_std)              # also std of S.mean[] is aprox. equal to (std of Age[])/root(50)
```
