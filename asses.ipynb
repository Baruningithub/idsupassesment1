```python
#1 Aprox derivative
def f1(x):
  return x*x
def difference_quotient(f, x, h=0.001):
    return (f(x + h) - f(x)) / h
difference_quotient(f1,2,0.001)
x=[i/10 for i in range(-50,50)]
y=[difference_quotient(f1,i,0.00001) for i in x]
y1=[2*i for i in x]
import matplotlib.pyplot as plt
plt.plot(x,y,'rx',label='approximation')
plt.plot(x,y1,'b+',label='original')
plt.legend()
plt.show()
```

```python
#2 Aprtial derivative and gradient descent
def f1(l):
  return (l[0]**2)*l[1]
def partial_difference_quotient(f, v, i, h):
    w = [v_j + (h if j == i else 0)
         for j, v_j in enumerate(v)]
    return (f(w) - f(v)) / h
def estimate_gradient(f, v, h=0.0001):
    return [partial_difference_quotient(f, v, i, h) 
            for i in range(len(v))]
estimate_gradient(f1, [2,3])
```

```python
#3 minimising derivative of x^2
def derivative_of_x_square(x):
  return 2*x
target_x=10
learning_rate=0.01
iterations=5
for i in range(iterations):
  target_x=target_x-(derivative_of_x_square(target_x))*learning_rate
print(target_x)
```

```python
#3 x2+y^2
def derivative_of_x_square_plus_y_square(x,y):
  return [2*x,2*y]
target_x,target_y=10,10
learning_rate=0.1
iterations=5
for i in range(iterations):
  target_x=target_x-(2*target_x*learning_rate)
  target_y=target_y-(2*target_y*learning_rate)
  print(i,target_x,target_y)
```

```python
#3 x1^2+x2^2+x3^2+ . . . +xn^2
def step(v, gradient, LearningRate):
    return [v_i + LearningRate * gradient_i
            for v_i, gradient_i in zip(v, gradient)]
def sum_of_squares_gradient(v):
    return [2 * v_i for v_i in v]
v=[1,1,1]
for i in range(10):
  gradient=sum_of_squares_gradient(v)
  v=step(v,gradient,-0.01)
  print(i,v)
```

```python
#5
def least_squares_fit(x,y):
    slope = correlation(x, y) * std(y) / std(x)
    intercept = mean(y) - slope * mean(x)
    return intercept, slope
```

```python
#6
def sum_of_sqerrors(intercept, slope, x, y) -> float:
    return sum(error(intercept, slope, x_i, y_i) ** 2
               for x_i, y_i in zip(x, y))
def total_sum_of_squares(y):
    return sum(v ** 2 for v in deviation(y))
def r_squared(intercept,slope, x, y) -> float:
    return 1.0 - (sum_of_sqerrors(intercept, slope, x, y) /
                  total_sum_of_squares(y))
intercept, slope = least_squares_fit(x,y)
rsq = r_squared(intercept,slope, x, y)
print(rsq)
```
